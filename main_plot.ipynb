```python
import pandas as pd
import json
from training_data.utils.build_basic_dict import build_basic_dict
import matplotlib.pyplot as plt
import os
```

```python
# load data
basic_df = build_basic_dict()
```

Basic Plots:

```python
## plot length of string

# prepare data
basic_df['str_len'] = basic_df['text'].apply(lambda x: len(x))

# set plot parameter
plot = basic_df.plot(x="goal", y="str_len")
plot.set_xticks(range(1,18,1))

# show
plt.show()
```

```python
## plot amount of words

# prepare data
basic_df['word_count'] = basic_df['text'].apply(lambda x: len(x.split(" ")))

# set plot parameter
plot = basic_df.plot(x="goal", y="word_count", legend=None)
plot.set_xticks(range(1,18,1))
plt.ylabel('amount words')

#show
plt.show()
```

```python
## plot amount of papers per sdg

# prepare data
scholar_data = pd.read_csv("files/scholar_data.csv")
scholar_data = scholar_data[scholar_data['pdf_text'].notna()]
paper_count = scholar_data.groupby('goal').count().reset_index()

# set plot parameter
plot = paper_count.plot(x='goal',y='pdf_text', legend=None)
plot.set_xticks(range(1,18,1))
plt.ylabel('amount papers')
plt.ylim((0,7))

# shot
plt.show()
```
